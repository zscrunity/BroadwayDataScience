```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Data/Salary_Data.csv' , usecols = ['Years of Experience' , 'Salary'])
dataset = dataset.dropna()
dataset
```

```python
X = dataset.iloc[:,0:-1].values
X
```

```python
Y = dataset.iloc[:,1:].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 1/3 , random_state = 10)
```

```python
total_len = (len(X))
print("All total length = ",total_len)
print("X_train length = ", len(X_train))
print("X_test length = ", len(X_test))
```

```python
# Training Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
# Testing Model
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]])

# Predict the Salary
predicted_salary = model.predict(exp)[0][0]
# print(predicted_salary)
print("Your predicted salary = " , predicted_salary)
```

```python
X_test.tolist()
```

```python
# For Testing Data
predicted_salary = model.predict(X_test)
for i in range(len(X_test)):
  print("Years of Experience : ", X_test.tolist()[i][0] , "\tSalary : " , predicted_salary[i][0])
```

```python
plt.scatter(X_train , Y_train , color = 'red')
plt.plot(X_train , model.predict(X_train) , color = 'blue')
plt.show()
```

```python
plt.scatter(X_test , Y_test , color = 'red')
plt.plot(X_train , model.predict(X_train) , color = 'blue')
plt.show()
```
