```python
# Regression between BMI and Expense
```

```python
df = pd.read_csv('../Data/insurance.csv' , usecols = ['bmi' , 'expenses'])
df
```

```python
X = df.iloc[:,0:-1].values
X
```

```python
Y = df.iloc[:,1:].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 10)
```

```python
total_len = (len(X))
print("All total length = ",total_len)
print("X_train length = ", len(X_train))
print("X_test length = ", len(X_test))
```

```python
# Training Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
# Testing Model
bmi = float(input("Enter BMI = "))
bmi = np.array([[bmi]])

# Predict the Salary
predicted_expenses = model.predict(bmi)[0][0]
print("Your expenses = " , predicted_expenses)
```

```python
# For Testing Data
predicted_expenses = model.predict(X_test)
for i in range(len(X_test)):
  print("BMI : ", X_test.tolist()[i][0] , "\tExpenses : " , predicted_expenses[i][0])
```

```python
plt.scatter(X_train , Y_train , color = 'red')
plt.plot(X_train , model.predict(X_train) , color = 'blue')
plt.show()
```

```python
plt.scatter(X_test , Y_test , color = 'red')
plt.plot(X_train , model.predict(X_train) , color = 'blue')
plt.show()
```
